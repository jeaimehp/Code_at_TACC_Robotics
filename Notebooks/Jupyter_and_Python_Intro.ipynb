## Conditionals (If Else)

Sometimes in a program you may need it to make a decision to do something based on a variable's value. These decisions are called *"Conditionals"* and use the syntax:

<b>
if expression:

    statement(s)

else:

    statement(s)
</b>  

Here is an example that tells you if a number is greater than 5. Change the number value of the variable *test* to try some other values.     

```python
test=3

if (test > 5):
    print (test,"is greater than 5")
else:
    print (test,"is lesser than 5")
    
```

## Loops (For and While)

You may want to check for a value or run a statement over, and over again. This type of program would be called a loop. 

### For Loop
Here is an example where we will test all the numbers between 0 and 15 to see if they are greater than 5.

```python
for test in range(15):
    if (test > 5):
        print (test,"is greater than 5")
    else:
        print (test,"is lesser than 5")
```

### While Loop
The while loop can be used until a certain condition is met and then stops.

Here is an example where we will test all the numbers from 0 until we hit one that is greater than 5.

```python
test=0
while (test < 5):
    print (test,"is lesser than 5, I'll keep trying")
    test = test+1
```

## Functions

Functions allow you take  statements, variables, conditionals, and loops and use them together as a group. That group can be used over and over again instead of typing all of them repeatedly. 

This example shows how we can check different numbers simply.

```python
def lesser_than (test):
    counter=0
    while (counter < test):
        print (counter,"is lesser than",test,"I'll keep trying")
        counter = counter+1
```

Now we can use the function to test numbers as we wish!

```python
lesser_than(3)
```

```python
lesser_than(10)
```
